```python
for col in df.select_dtypes(include=['object']).columns:
  df[col] = df[col].astype(str)
  label_encoder=LabelEncoder()
  label_encoder.fit(df[col])
  df[col]=label_encoder.transform(df[col])
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
scalar=StandardScaler()
df=scalar.fit_transform(df)
df
```

```python
pca=PCA(n_components=6)
dfs=pca.fit_transform(df)
dfs
```

```python
pca_df=pd.DataFrame(data=dfs,columns=[f'pc{i+1}'for i in range(6)])
df = pd.read_csv('/content/Supermarket_sales_prediction.csv')
pca_df['Item_Outlet_Sales']=df['Item_Outlet_Sales']
pca_df
```

```python
x=pca_df.drop('Item_Outlet_Sales',axis=1)
y=pca_df['Item_Outlet_Sales']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=50)
x_train,x_test,y_train,y_test
```

```python
LR_model=LinearRegression()
LR_model.fit(x_train,y_train)
y_pred=LR_model.predict(x_test)

r2_score_value=r2_score(y_test,y_pred)
print(r2_score_value)
```

```python
RF_model=RandomForestRegressor()
RF_model.fit(x_train,y_train)
y_pred=RF_model.predict(x_test)

r2_score_value=r2_score(y_test,y_pred)
print(r2_score_value)
```

```python
GD_model=GradientBoostingRegressor()
GD_model.fit(x_train,y_train)
y_pred=GD_model.predict(x_test)

r2_score_value=r2_score(y_test,y_pred)
print(r2_score_value)
```

```python
import lightgbm as lgb
LGB_model = lgb.LGBMRegressor()
LGB_model.fit(x_train, y_train)
y_pred=LGB_model.predict(x_test)

r2_score_value=r2_score(y_test,y_pred)
print(r2_score_value)
```
